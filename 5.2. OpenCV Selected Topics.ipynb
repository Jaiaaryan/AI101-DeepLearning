### 5.2.1 Canny Edge Detection

```python
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('img/dog.jpeg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img2 = cv2.Canny(img_gray,100,180)
plt.imshow(img2,cmap='gray')
plt.axis('off')
```

### 5.2.1 Putting Bounding Box

```python
import matplotlib.pyplot as plt
import numpy as npfontcolor = (255,255,255)

font = cv2.FONT_HERSHEY_SIMPLEX
fontscale = 1
fontcolor = (255,255,255)
fontcolor2 = (0,255,0)
fontwidth=2

img = cv2.imread('img/dog.jpeg')
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.rectangle(img2,(180,10),(500,300),(0,255,0),3)
cv2.rectangle(img2,(180,265),(270,300),(0,255,0),-1)
cv2.putText(img2,'Dog',(190,290),font,fontscale,fontcolor,fontwidth)
cv2.putText(img2,'0.47',(275,290),font,fontscale,fontcolor2,fontwidth)

plt.figure(figsize=(15,7))
plt.imshow(img2)
plt.axis('off')

```
